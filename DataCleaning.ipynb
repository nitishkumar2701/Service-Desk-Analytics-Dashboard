```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/Technical Support Dataset.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.describe()
```

```python
df.head()
```

```python
df.info()
```

```python
df['Survey results'] = df['Survey results'].fillna(df['Survey results'].mean()).round()
```

```python
df['Survey results'].info()
```

```python
df['Agent interactions'] = df['Agent interactions'].fillna(df['Agent interactions'].mean()).round()
```

```python
df['Agent interactions']
```

```python
df['First response time']
```

```python
df['First response time'] = pd.to_datetime(df['First response time'])
df['First response time'] = df['First response time'].fillna(df['First response time'].median())
```

```python
df['Resolution time'] = pd.to_datetime(df['Resolution time'])
df['Resolution time'] = df['Resolution time'].interpolate()
```

```python
df['Resolution time']
```

```python
df['Close time']
```

```python
df['Close time'] = pd.to_datetime(df['Close time'])
df['Close time'] = df['Close time'].interpolate()
```

```python
df.to_excel('CustSupportTickets.xlsx', index=False)
```
